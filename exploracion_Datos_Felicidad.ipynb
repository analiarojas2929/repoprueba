```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

df = pd.read_csv('2015.csv')  # Aseg√∫rate que el archivo est√© en tu directorio

print("üîç Primeras filas del dataset:")
display(df.head())

print("üìã Columnas disponibles:")
print(df.columns)

# Variables con nombres en espa√±ol para mostrar
variables = {
    'Puntaje de Felicidad': 'Happiness Score',
    'PIB per c√°pita': 'Economy (GDP per Capita)',
    'Esperanza de vida saludable': 'Health (Life Expectancy)',
    'Libertad para tomar decisiones de vida': 'Freedom'
}

def graficar_histograma_boxplot(df, columna, color='skyblue'):
    fig, axs = plt.subplots(nrows=2, figsize=(10, 6), gridspec_kw={'height_ratios': [3, 1]})

    axs[0].hist(df[columna].dropna(), bins=20, color=color, edgecolor='black', alpha=0.7)
    axs[0].set_title(f'Histograma de {columna}')
    axs[0].set_ylabel('Frecuencia')

    sns.boxplot(x=df[columna], ax=axs[1], color=color)
    axs[1].set_title(f'Boxplot de {columna}')
    axs[1].set_xlabel(columna)

    plt.tight_layout()
    plt.show()

for nombre_espanol, columna in variables.items():
    print(f"üîç An√°lisis de la variable: {nombre_espanol}")
    graficar_histograma_boxplot(df, columna)
```
