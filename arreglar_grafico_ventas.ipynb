```python
import pandas as pd
import matplotlib.pyplot as plt
from io import StringIO  # Importacin necesaria para leer desde texto

# Datos de ejemplo en formato CSV (como string)
data = """
fecha,ventas
2023-01-01,200
2023-01-02,220
2023-01-03,210
2023-01-04,205
2023-01-05,215
2023-01-06,225
"""

# Cargar los datos en un DataFrame de Pandas usando StringIO
df = pd.read_csv(StringIO(data))

# Convertir la columna 'fecha' a formato datetime correctamente
df['fecha'] = pd.to_datetime(df['fecha'])

#  Grfico de lneas de las ventas
plt.figure(figsize=(10, 6))
plt.plot(df['fecha'], df['ventas'], marker='o', linestyle='-')
plt.title("Ventas Diarias")
plt.xlabel("Fecha")
plt.ylabel("Ventas")
plt.grid(True)
plt.tight_layout()
plt.show()

#  Grfico de dispersin
plt.figure(figsize=(10, 6))
plt.scatter(df['fecha'], df['ventas'], color='orange')
plt.title("Ventas Diarias - Dispersin")
plt.xlabel("Fecha")
plt.ylabel("Ventas")
plt.grid(True)
plt.tight_layout()
plt.show()

#  Histograma de las ventas
plt.figure(figsize=(8, 5))
plt.hist(df['ventas'], bins=5, color='skyblue', edgecolor='black')
plt.title("Distribucin de Ventas")
plt.xlabel("Ventas")
plt.ylabel("Frecuencia")
plt.grid(True)
plt.tight_layout()
plt.show()
```
